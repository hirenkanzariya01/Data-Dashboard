```python
def getNumaricColums(a):
    isNumaric = []
    if df[a].dtype == 'floar64' or df[a].dtype == 'int64':
        isNumaric.append(a)
    return isNumaric
```

```python
for i in df.columns:
    getNumaricColums(i)
```
